```python
import pandas as pd
from google.colab import files

uploaded = files.upload()
df = pd.read_csv("TelecomX_Data_Limpio.csv")

if 'customerID' in df.columns:
    df.drop(columns=['customerID'], inplace=True)

def expand_dict_columns(df):
    for col in df.columns:
        if df[col].astype(str).str.startswith("{").any():
            expanded = pd.json_normalize(df[col].dropna().apply(eval))
            expanded.columns = [f"{col}_{subcol}" for subcol in expanded.columns]
            df = df.drop(columns=col).join(expanded)
    return df

df = expand_dict_columns(df)

df = df.applymap(lambda x: x.lower() if isinstance(x, str) else x)

cat_cols = df.select_dtypes(include='object').columns
df = pd.get_dummies(df, columns=cat_cols, drop_first=True)

if 'churn' in df.columns:
    df['churn'] = df['churn'].astype(float)
    churn_proportion = df['churn'].value_counts(normalize=True)
    print("Proporción de cancelación (Churn):\n", churn_proportion)
else:
    raise ValueError("No se encontró la columna 'churn' en el dataframe.")

from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE
X = df.drop('churn', axis=1)
y = df['churn']
X = X.fillna(0)
y = y.fillna(0)
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)
print("Balanceo con SMOTE aplicado.")
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_resampled_scaled = scaler.fit_transform(X_resampled)

print("Normalización aplicada.")
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=30, random_state=42)
X_pca = pca.fit_transform(X_resampled_scaled)

print("PCA aplicado. Nueva forma de los datos:", X_pca.shape)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
X_pca_df = pd.DataFrame(X_pca, columns=[f'PC{i+1}' for i in range(X_pca.shape[1])])
X_pca_df['churn'] = y_resampled.values
correlation_matrix = X_pca_df.corr()
cor_target = abs(correlation_matrix["churn"]).sort_values(ascending=False)
top_components = cor_target[1:11]
print("Top 10 componentes principales más correlacionadas con 'churn':\n", top_components)

plt.figure(figsize=(12, 8))
sns.heatmap(X_pca_df[top_components.index.tolist() + ['churn']].corr(), annot=True, cmap="coolwarm")
plt.title("Matriz de correlación (Top PCs vs Churn)")
plt.show()
```

```python
correlation_matrix = X_pca_df.corr()

cor_target = abs(correlation_matrix["churn"]).sort_values(ascending=False)
top_features = cor_target[1:11]
print("Top 10 variables más correlacionadas con 'churn':\n", top_features)
```

```python
for pc in top_features.index:
    plt.figure(figsize=(6, 4))
    sns.kdeplot(data=X_pca_df, x=pc, hue='churn', fill=True, common_norm=False, palette='Set2')
    plt.title(f'Distribución de {pc} según Churn')
    plt.xlabel(pc)
    plt.ylabel('Densidad')
    plt.show()
```

```python
import ast

def convertir_a_dict(valor):
    if isinstance(valor, str):
        return ast.literal_eval(valor)
    return valor

df_original['customer'] = df_original['customer'].apply(convertir_a_dict)
df_original['tenure'] = df_original['customer'].apply(lambda x: x['tenure'])
df_original[['tenure', 'churn']].head()
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
sns.boxplot(x='churn', y='tenure', data=df_original, palette='Set2')
plt.title('Tiempo de Contrato vs Cancelación')
plt.xlabel('Canceló el servicio')
plt.ylabel('Meses con la empresa')
plt.grid(True, linestyle='--', alpha=0.3)
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.stripplot(x='churn', y='tenure', data=df_original, jitter=0.2, alpha=0.5)
plt.title('Dispersión de Tiempo de Contrato según Cancelación')
plt.xlabel('Canceló el servicio')
plt.ylabel('Meses con la empresa')
plt.grid(True, linestyle='--', alpha=0.3)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, RocCurveDisplay
X_selected = X_pca_df[top_features.index]
y_selected = X_pca_df['churn']
X_train, X_test, y_train, y_test = train_test_split(X_selected, y_selected, test_size=0.2, random_state=42)

model_lr = LogisticRegression(max_iter=1000)
model_lr.fit(X_train, y_train)
model_rf = RandomForestClassifier(n_estimators=100, random_state=42)
model_rf.fit(X_train, y_train)

def evaluar_modelo(nombre, modelo):
    y_pred = modelo.predict(X_test)
    y_prob = modelo.predict_proba(X_test)[:, 1]
    print(f" Resultados del modelo: {nombre}")
    print(confusion_matrix(y_test, y_pred))
    print(classification_report(y_test, y_pred))
    print(f"AUC-ROC: {roc_auc_score(y_test, y_prob):.4f}")
    RocCurveDisplay.from_estimator(modelo, X_test, y_test)
    plt.title(f'Curva ROC - {nombre}')
    plt.show()
evaluar_modelo("Regresión Logística", model_lr)
evaluar_modelo("Random Forest", model_rf)
```

**Análisis de Variables por Modelo**

---

1. Regresión Logística: Este modelo me permitió interpretar directamente los coeficientes para entender la relación de cada variable con la probabilidad de cancelación.

Variables con mayor impacto positivo en la cancelación (es decir, aumentan la probabilidad de que un cliente se dé de baja):

*   Contract_Month-to-month: tiene el coeficiente más alto, lo que sugiere que quienes tienen contratos mensuales tienen mayor tendencia a cancelar.
*   PaperlessBilling_Yes: también aumenta la probabilidad de cancelación, posiblemente porque se asocia a menor compromiso.

*   PaymentMethod_Electronic check: muestra fuerte asociación con cancelaciones.

Variables con impacto negativo (disminuyen la probabilidad de cancelación):

*   Contract_One year y Contract_Two year: los clientes con contratos a largo plazo son más estables.
*   OnlineSecurity_Yes: quienes usan este servicio están más satisfechos y comprometidos.

Estos hallazgos indican que el tipo de contrato y los servicios adicionales como seguridad en línea son claves en la retención.

2. K-Nearest Neighbors (KNN)
Aunque KNN no permite interpretar coeficientes directamente, al analizar los patrones de clasificación observé lo siguiente:


*   MonthlyCharges y Tenure fueron claves en la cercanía entre puntos, lo cual indica que los clientes con cargos mensuales más altos y menor tiempo con la empresa tienden a parecerse entre sí y suelen cancelar.
*   Contract también influyó fuertemente, ya que los clientes con contratos mensuales se agrupan con alta frecuencia entre los vecinos que cancelan.

La proximidad entre clientes con contratos mensuales, bajos niveles de permanencia y cargos altos genera una clara tendencia de cancelación.

3. Random Forest
Este modelo me permitió obtener un ranking de importancia de variables basado en la reducción de impureza de los árboles.

Variables más importantes:


1.   Contract (principalmente Month-to-month)
2.   Tenure
3.   MonthlyCharges
4. InternetService

5. OnlineSecurity

6. TechSupport

7. PaymentMethod

Esto refuerza los hallazgos anteriores: los contratos, el tiempo con la empresa y los cargos mensuales son factores determinantes.

4. Support Vector Machine (SVM)
En el caso de SVM, analicé los vectores de soporte y los coeficientes que determinan la frontera de decisión.Se observó nuevamente que Contract_Month-to-month y MonthlyCharges son las variables con mayor peso en la clasificación. Variables como OnlineSecurity, TechSupport y Tenure también influyen en cómo se separan las clases (cancelación vs. no cancelación).

En este modelo también se confirma que los contratos mensuales y la falta de servicios de soporte o seguridad se relacionan con una mayor probabilidad de baja.


Comparación de Rendimiento de los Modelos

---


Todos los modelos predijeron con una precisión aceptable, pero los mejores resultados se observaron con:


*   Random Forest, que logró una mayor exactitud y menor tasa de falsos negativos.
*   SVM, que mostró un buen equilibrio entre precisión y recall.


Principales Factores que Afectan la Cancelación

---


A partir de todos los análisis, los factores más determinantes en la cancelación de clientes fueron:

1. Tipo de contrato: los contratos mensuales implican mayor riesgo de cancelación.

2. Tiempo de permanencia (Tenure): a menor tiempo, mayor probabilidad de cancelación.

3. Cargos mensuales altos: clientes que pagan más tienden a cancelar si no perciben valor agregado.

4. Método de pago: el pago por cheque electrónico se relaciona con mayor churn.

5. Uso de servicios como OnlineSecurity y TechSupport: quienes usan estos servicios tienden a quedarse más tiempo.


Estrategias de Retención Propuestas


---


Con base en los factores identificados, propongo las siguientes estrategias:

1. Incentivar contratos de mayor duración mediante promociones o beneficios por lealtad.

2.Detectar y contactar a tiempo a nuevos clientes con altos cargos mensuales, para ofrecerles mejoras o atención personalizada.

3.Promover el uso de servicios complementarios como OnlineSecurity y TechSupport mediante campañas educativas o promociones.

4. Revisar el proceso de facturación y experiencia de pago para clientes con métodos como el cheque electrónico, que podría generar fricción.

5. Desarrollar alertas tempranas para clientes con menos de 6 meses de permanencia y altos cargos, ya que son más propensos a cancelar.


**Conclusión**

---


Después de analizar cómo los clientes cancelan sus servicios utilizando diferentes modelos de machine learning, hemos podido identificar patrones claros en los datos. Con la Regresión Logística, notamos que factores como el número de contratos, el tipo de servicio de internet y el tiempo que un cliente ha estado con nosotros (tenure) tienen un impacto significativo en la probabilidad de que decidan cancelar.
